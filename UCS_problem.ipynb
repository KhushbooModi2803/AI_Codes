```python
## UCS Code
from queue import PriorityQueue

graph = {
    'S': {'A': 1, 'B': 5, 'C': 15},
    'A': {'G': 10},
    'B': {'G': 5},
    'C': {'G': 5}
}

def is_goal(state, goal):
    return state == goal

def get_neighbors(state):
    return graph[state].keys()

def get_cost(state, neighbor):
    return graph[state][neighbor]

def uniform_cost_search(start, goal):
    frontier = PriorityQueue()
    frontier.put((0, start, [start]))  # Store path along with cost
    explored = set()

    while frontier:
        current_cost, state, path = frontier.get()

        if is_goal(state, goal):
            return path, current_cost

        if state in explored:
            continue

        explored.add(state)

        for neighbor in get_neighbors(state):
            new_cost = current_cost + get_cost(state, neighbor)
            new_path = path + [neighbor]
            if neighbor not in explored and neighbor not in frontier.queue:
                frontier.put((new_cost, neighbor, new_path))
            elif neighbor in frontier.queue:
                if new_cost < frontier.queue[neighbor][0]:
                    frontier.put((new_cost, neighbor, new_path))

    return None, None  # No solution found

# Example usage
start = 'S'
goal = 'G'

path, cost = uniform_cost_search(start, goal)

if path:
    print("Solution path:", path, "with cost:", cost)
else:
    print("No solution found.")
```
