```python
## BFS and DFS
from collections import deque

def get_neighbors(state):
    """Generate all possible next states from the current state."""
    neighbors = []
    index_empty = state.index('_')
    for i in range(3):  # Assuming 3 slots
        if i != index_empty:
            # Swap the empty slot with each of the blocks
            new_state = list(state)
            new_state[index_empty], new_state[i] = new_state[i], new_state[index_empty]
            neighbors.append(''.join(new_state))
    return neighbors



def bfs(start, goal):
    """Breadth-First Search from start state to goal state."""
    visited = set()
    queue = deque([(start, [start])])
    
    while queue:
        current_state, path = queue.popleft()
        if current_state == goal:
            return path
        
        for neighbor in get_neighbors(current_state):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
                
def dfs(start, goal, path=[], visited=set()):
    """Depth-First Search from start state to goal state."""
    if start == goal:
        return path + [start]
    
    visited.add(start)
    for neighbor in get_neighbors(start):
        if neighbor not in visited:
            result = dfs(neighbor, goal, path + [start], visited)
            if result:
                return result

# Define initial and goal states
initial_state = 'ABC_'
goal_state = 'ACB_'

# Solve using BFS
bfs_path = bfs(initial_state, goal_state)
print("BFS Path:", bfs_path)

```
